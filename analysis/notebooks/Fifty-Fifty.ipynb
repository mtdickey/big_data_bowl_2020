### Background 

A common phrase used to describe a pass that has an equal chance to be caught by either team is known as a "50-50" ball.  Often,  they are "jump ball" scenarios where the quarterback throws a high-arcing pass between his intended receiver and their defender.

These passes present an enormous opportunity for the defense, *if* they can take advantage of it.

For example, in Week 16 of the 2018 NFL regular season the Baltimore Ravens were fighting to win their division and make the playoffs when they faced the L. A. Chargers.  On the first offensive play of the game, Philip Rivers threw a deep pass along the sideline to Mike Williams.  Brandon Carr made an outstanding play to wrestle the ball away from Williams and gave the Ravens great field position to start their first possession.

![williams](img/williams_50-50.gif)

This play set the tone for the game and the Ravens won convincingly with a score of 22 to 10, but not before they stopped the Chargers late push with another win on a 50-50 ball.

![benjamin](img/benjamin_50-50.gif)

The 2018 Ravens went on to win their division by half a game and made the playoffs.  A large contributor to their success that season was their defense, which allowed the 2nd fewest points and the 2nd fewest yards per pass attempt in the regular season.




### Method


This analysis uses logistic regression to estimate the likelihood of a "coverage success" on a pass play.

"Coverage success" is defined by the following events:
  1. Coverage wins
     - Incomplete pass
     - Intercepted pass
     - Offensive pass interference penalty accepted (not including events where there were offsetting penalties)
  2. Coverage losses
     - Completion
     - Defensive coverage penalty accepted
        - Defensive pass interference
        - Defensive holding
        - Illegal contact

Several factors were analyzed to determine their effect on coverage success, including:
  - Player/ball tracking factors:
     - Distance of pass thrown downfield (beyond the line-of-scrimmage)
     - Distance from receiver to closest defender, at the time that the pass arrives
     - Distance from receiver to the ball, at the time that the pass arrives
  - Player characteristics
     - Difference between receiver and closest defender height
